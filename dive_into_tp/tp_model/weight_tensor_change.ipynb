# Changing core tensor in tucker decomposition

```python
from tensorly.decomposition import tucker
from tensorly import tucker_to_tensor
```

```python
core, factors = tucker(A, rank=[4, 7, 7])
```

### Multiply by a scalar value

```python
NUM = 0.9
core_1 = core*NUM
```

```python
v = tucker_to_tensor((core_1, factors))
```

```python
v.sum(axis=0)
```

```python
for c in v:
 plt.imshow(c, cmap="magma")
 plt.show()
```

```python
NUM = 10.9
core_2 = core*NUM
```

```python
v = tucker_to_tensor((core_1, factors))
for c in v:
 plt.imshow(c, cmap="magma")
 plt.show()
```

## Changing weight tensor

```python
NUM = 0.5
factors_1 = [f*NUM for f in factors]
```

```python
v = tucker_to_tensor((core, factors_1))
for c in v:
 plt.imshow(c, cmap="magma")
 plt.show()
```

```python
np.sum(v, axis=0)
```
